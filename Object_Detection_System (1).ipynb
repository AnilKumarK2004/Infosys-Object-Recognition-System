```python
%%writefile app.py
import streamlit as st
import torch
import numpy as np
import cv2
from PIL import Image
from ultralytics import YOLO
import pandas as pd
import os
from torchvision import transforms

# Load the trained YOLOv8 model
model = YOLO('runs/detect/train/weights/best.pt')  # Path to the best model weights

# Function to perform inference on an uploaded image
def inference(image):
    image = np.array(image)  # Convert image to NumPy array
    results = model(image)   # Perform inference using YOLOv8 model
    return results

# Function to perform image augmentation
def augment_images(image):
    augmentations = transforms.Compose([
        transforms.RandomHorizontalFlip(p=0.5),  # 50% chance of flipping
        transforms.RandomRotation(degrees=15),   # Rotate images by up to 15 degrees
        transforms.ColorJitter(brightness=0.2, contrast=0.2, saturation=0.2, hue=0.1),  # Random color changes
        transforms.RandomResizedCrop(size=(640, 640), scale=(0.8, 1.0)),  # Randomly crop the image
    ])

    # Generate augmented images
    augmented_images = []
    for _ in range(5):  # Create 5 augmented images
        augmented_image = augmentations(image)
        augmented_images.append(augmented_image)

    return augmented_images

# Function to train the model with uploaded images
def train_model(image_paths, class_names):
    # Prepare a dataset for training
    # Create a directory for training images
    os.makedirs("train_images", exist_ok=True)

    # Save the original images and create augmented images
    for img_path in image_paths:
        image = Image.open(img_path)
        image.save(f"train_images/{os.path.basename(img_path)}")  # Save original image

        # Generate and save augmented images
        augmented_images = augment_images(image)
        for i, aug_img in enumerate(augmented_images):
            aug_img.save(f"train_images/aug_{os.path.basename(img_path).split('.')[0]}_{i}.jpg")

    # Prepare dataset.yaml for YOLO training
    data_yaml = f"""
    train: train_images  # Path to training images
    val: train_images    # Path to validation images (use original images)
    nc: {len(class_names)}  # Number of classes
    names: {class_names}  # List of class names
    """

    # Save the YAML file
    with open('dataset.yaml', 'w') as f:
        f.write(data_yaml)

    # Train the model
    model.train(data='dataset.yaml', epochs=50, imgsz=640)

# Function to draw bounding boxes on the image
def draw_bounding_boxes(results, image):
    img = np.array(image)

    for r in results[0].boxes:
        box = r.xyxy[0]  # Get the bounding box coordinates (x1, y1, x2, y2)
        conf = r.conf[0].item()  # Confidence score
        label = model.names[int(r.cls[0])]  # Object class

        # Draw bounding box on the image
        cv2.rectangle(img, (int(box[0]), int(box[1])), (int(box[2]), int(box[3])), (0, 255, 0), 2)
        cv2.putText(img, f'{label} {conf:.2f}', (int(box[0]), int(box[1]) - 10),
                    cv2.FONT_HERSHEY_SIMPLEX, 0.9, (36, 255, 12), 2)

    return img

# Streamlit app for inference and training
st.title("YOLOv8 Object Detection and Training")

# Section for training the model
st.header("Train the Model with New Images")
uploaded_train_files = st.file_uploader("Upload up to 30 images for training", type=['jpg', 'jpeg', 'png'], accept_multiple_files=True)

# Input for class names
num_classes = st.number_input("Number of classes", min_value=1, max_value=10, value=1)
class_names = st.text_area("Enter class names (comma-separated)", "class_name1,class_name2")

if uploaded_train_files and len(uploaded_train_files) <= 30:
    train_image_paths = []  # List to store paths of uploaded training images

    for uploaded_file in uploaded_train_files:
        # Save uploaded images to a directory
        img_path = f"train_images/{uploaded_file.name}"
        image = Image.open(uploaded_file)
        image.save(img_path)
        train_image_paths.append(img_path)

    if st.button("Train Model"):
        # Split the class names into a list and strip whitespace
        class_names_list = [name.strip() for name in class_names.split(",") if name.strip()]
        train_model(train_image_paths, class_names_list)  # Train the model with uploaded images
        st.success("Model training completed!")

# Section for performing inference on an unseen image
st.header("Perform Inference on an Unseen Image")
uploaded_infer_file = st.file_uploader("Upload an image for inference", type=['jpg', 'jpeg', 'png'])

if uploaded_infer_file:
    # Load the image using PIL
    image = Image.open(uploaded_infer_file)
    st.image(image, caption="Uploaded Image for Inference", use_column_width=True)

    # Perform inference when the 'Detect Objects' button is clicked
    if st.button("Detect Objects"):
        results = inference(image)  # Run YOLOv8 inference on the uploaded image

        # Display detection results
        if results:  # Check if there are any results
            st.write("Detected Objects:")
            for result in results:  # Iterate through the results
                boxes = result.boxes  # Get the boxes attribute

                # Prepare a list for storing detection information
                detection_info = []

                for box in boxes:
                    box_coords = box.xyxy[0].tolist()  # Convert bounding box coordinates to list
                    conf = box.conf[0].item()  # Confidence score
                    label = model.names[int(box.cls[0])]  # Get class label

                    # Append the detection information
                    detection_info.append({
                        'x1': box_coords[0],
                        'y1': box_coords[1],
                        'x2': box_coords[2],
                        'y2': box_coords[3],
                        'confidence': conf,
                        'label': label
                    })

                # Display the detection information in a dataframe format
                if detection_info:  # Check if there are detections to display
                    df = pd.DataFrame(detection_info)
                    st.write(df)  # Display the dataframe
                else:
                    st.write("No objects detected.")

        # Draw bounding boxes on the image
        img_with_boxes = draw_bounding_boxes(results, image)

        # Convert NumPy array back to Image and display it
        st.image(img_with_boxes, caption="Image with Bounding Boxes", use_column_width=True)


```
